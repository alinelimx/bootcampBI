DESAFIO 1

Peça o número de páginas lidas em um dia:

A: 50 ou mais → Leitor voraz

B: 30 a 49 → Ótimo ritmo

C: 10 a 29 → Leitura moderada

D: menos de 10 → Leitura leve

```python
def controle_leitura(numPagsLidas):
  if numPagsLidas >= 50:
    print("parabens!! voce é um leitor voraz")
  elif numPagsLidas >= 30 and numPagsLidas <= 49:
    print("continue assim!! voce está tendo um otimo ritmo")
  elif numPagsLidas >= 10 and numPagsLidas <= 29:
    print("seu ritmo de leitura está moderado")
  else:
    print("voce está tendo um ritmo de leitura leve")

numPagsLidas = int(input("digite a quantidade de paginas que voce leu em um dia: "))
controle_leitura(numPagsLidas)
```

DESAFIO 2

Solicite uma string e substitua todas as vogais por * . Exemplo: "Olá mundo" → "*l* m*nd*".

```python
def substituindoVogais(frase):
  vogais = ['a', 'e', 'i', 'o', 'u']
  for vogal in vogais:
    frase = frase.replace(vogal, '*')
  print(frase)

frase = input("digite uma frase: ")
substituindoVogais(frase)
```

DESAFIO 3

O computador escolhe um número entre 1 e 50.

O usuário tem apenas 5 tentativas para adivinhar.
Se acertar, parabéns! Se não, o jogo acaba e revela o número.

```python
import random

def adivinharNum():
  num = random.randint(1, 50)
  controleTentativas = 0
  for i in range(5):
    controleTentativas += 1
    try:
      tentativa = int(input("digite um numero entre 1 e 50: "))
    except ValueError:
        print("digite apenas numeros!!")
        continue

    if tentativa < 1 or tentativa > 50:
          print("so vale numeros entre 1 e 50")
          continue

    if tentativa == num:
      print("parabens voce acertou!! :)")
      break
    elif tentativa > num:
      print("tente um numero menor")
    else:
      print("tente um numero maior")
  else:
    print(f"voce perdeu :/ o numero era {num}")

adivinharNum()
```

DESAFIO 4

Permita:

Adicionar um alimento (nome, calorias)

Registrar refeições do dia

Ver total de calorias consumidas

Ver histórico alimentar

Sair

```python
alimentos = []
refeicoes_dia = []

def menu():
    print("\n=== MENU ===")
    print("1 - adicionar alimento")
    print("2 - registrar refeicao")
    print("3 - ver total de calorias")
    print("4 - ver historico alimentar")
    print("5 - sair")

while True:
    menu()
    opcao = int(input("escolha uma opcao: "))

    if opcao == 1:
        nome = input("nome do alimento: ")
        calorias = int(input("quantidade de calorias: "))
        alimentos.append({"nome": nome,
                          "calorias": calorias}) #cria um dicionario com as chaves nome e calorias e adiciona na lista alimentos
        print(f"alimento adicionado")
        print('-='*15)

    elif opcao == 2:
        if not alimentos: #serve pra verificar se ta vazio
            print("nenhum alimento foi cadastrado pra poder registrar a refeicao")
        else:
            print("\nalimentos disponiveis:")
            for posicao, alimento in enumerate(alimentos): # percorre a lista mostrando ao mesmo tempo a posicao e o alimento na lista
                print(f"{posicao} - {alimento['nome']} ({alimento['calorias']} cal)")

            itens = input("digite os itens dos alimentos consumidos separando eles por um espaço: ")
            itens = itens.split()

            refeicao = [] #lista q vai guardar os alimentos consumidos numa refeicao (cada refeicao seria uma lista de alimentos)
            for item in itens:
                if item.isdigit() and int(item) < len(alimentos): # ve se é um numero e se ta dentro da lista mesmo
                    refeicao.append(alimentos[int(item)]) #coloca o alimento correspondente ao item e bota na lista refeicao
            refeicoes_dia.append(refeicao) # adiciona a refeicao a lista de refeicoes do dia
            print("refeicao registrada com sucesso")
        print('-='*15)

    elif opcao == 3:
        total = 0
        for refeicao in refeicoes_dia: #percorre todas as refeicoes registradas
            for alimento in refeicao: #pra cada alimento em cada refeicao soma as calorias
                total += alimento['calorias']
        print(f"total de calorias consumidas no dia: {total} cal")
        print('-='*15)

    elif opcao == 4:
        if not refeicoes_dia:
            print("nenhuma refeicao registrada")
        else:
          #mostra todas as refeicoes registradas e o q foi comido em cada uma
          for posicao, refeicao in enumerate(refeicoes_dia):
            print(f"\nrefeicao {posicao + 1}:") #mostra o numero da refeição (usei o posicao + 1 para começar do 1)
            for alimento in refeicao:
              print(f" - {alimento['nome']} ({alimento['calorias']} cal)")
        print('-='*15)

    elif opcao == 5:
        print("saindo do programa")
        break

    else:
        print("opcao invalida! tente novamente apenas com numeros entre 1 e 5")
```

DESAFIO 5

Peça ao usuário a distância percorrida (km) e o total de combustível gasto (litros).
Calcule o consumo médio:
Consumo = distância / litros

Classifique o consumo:

Menor que 8 km/l: Consumo alto

Entre 8 e 12 km/l: Consumo médio

Maior que 12 km/l: Consumo eficiente

```python
def consumo(distancia, combustivel):
  consumoMedio = distancia / combustivel

  if consumoMedio < 8:
    print(f"seu consumo é de {consumoMedio}, que se classifica como consumo alto")
  elif consumoMedio >= 8 and consumoMedio <= 12:
    print(f"seu consumo é de {consumoMedio}, que se classifica como consumo medio")
  else:
    print(f"seu consumo é de {consumoMedio}, que se classifica como consumo eficiente")

distancia = float(input("digite a distancia percorrida em quilometros: "))
combustivel = float(input("digite o total de combustivel gasto em litros: "))

consumo(distancia, combustivel)
```
